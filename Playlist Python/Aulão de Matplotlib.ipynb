## Conceitos Gerais
- **Figure**: É a área total da plotagem que pode conter um ou mais gráficos (Axes).
- **Axes**: É o que normalmente pensamos como um gráfico individual dentro da Figure.
- **Axis (eixos)**: São as linhas numéricas como objetos que cuidam dos limites dos gráficos.
- **Artist**: Tudo o que se pode ver na Figure é um Artist, como objetos de Texto, Linha2D, entre outros.

## Começando com Pyplot
Pyplot é um módulo do Matplotlib que fornece funções simples para adicionar elementos como linhas, imagens e texto aos eixos atuais na figura atual. Na maioria dos casos iremos utilizar o Matplotlib juntamente ao módulo pyplot.

Para demonstrar como se cria um gráfico, observe o código abaixo:

```python
# Importando as bibliotecas
import matplotlib.pyplot as plt
import numpy as np

# Exemplo de dados
x = np.arange(0, 10, 0.2)
y = np.sin(x)

# Criando o gráfico, com título e nome dos eixos
plt.figure()
plt.plot(x, y)
plt.title("Seno de X")
plt.xlabel('x')
plt.ylabel("sin(x)")
plt.xlim(0, 9.8)
plt.ylim(-1, 1)
plt.show()
```

Vamos alterar o tamanho da figura e o estilo da linha?

```python
# Plotando o gráfico e adicionando a cor verde e tracejado
plt.figure(figsize = (4, 3))
plt.plot(x, y, 'g--')
plt.title("Seno de X")
plt.xlabel('x')
plt.ylabel("sin(x)")
plt.xlim(0, 9.8)
plt.ylim(-1, 1)
plt.show()
```

Agora, vamos personalizar um pouco esse gráfico. Deixar mais bonito.

```python
# Colocando o estilo fivethirtyeight
plt.style.use('fivethirtyeight')
plt.figure(figsize = (4, 3))
plt.plot(x, y, 'g--')
plt.title("Seno de X")
plt.xlabel('x')
plt.ylabel("sin(x)")
plt.show()
```

Ficou beeem mais bonito. Agora, vamos fazer múltiplos gráficos!

```python
# Adicionando 2 subplots (verde -> seno & vermelho -> cosseno)
plt.figure(figsize = (16, 9))

plt.subplot(2, 2, 1)
plt.plot(x, y, 'g--', label = "sen(x)")
plt.legend()
plt.title("Sen(x)")

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'r:', label = 'cos(x)')
plt.legend()
plt.title("Cos(x)")

plt.tight_layout()
plt.show()
```

Com isso, podemos avançar para os outros tipos de gráficos, pois a personalização é idêntica.

## Gráfico de Barras
O gráfico de barras é muito utilizado para mostrar a frequência de ocorrências de valores de uma variável categórica. Abaixo, está seu exemplo de código.

```python
# Criando os dados
marcas = ['Marca A', 'Marca B', 'Marca C']
vendas_em_M = [10, 20, 15]

# Plotando o gráfico
plt.figure(figsize = (16, 9))
plt.bar(marcas, vendas_em_M)
plt.title("Vendas das marcas em Milhões")
plt.xlabel('Marca')
plt.ylabel("Vendas (em milhões)")
plt.show()
```

## Histograma
Os histogramas são frequentemente utilizados para representar a distribuição das variáveis. Para cria-lo, execute o código abaixo.

```python
# Criando os dados
dados = np.random.randn(100000)

# Plotando o gráfico
plt.figure(figsize = (16, 9))
plt.hist(dados, bins = 30)
plt.title("Distribuição dos dados")
plt.show()
```

## Gráfico de Dispersão (scatter plot)
O gráfico de dispersão pode ser muito utilizado para identificar possíveis relações entre dados. Para cria-lo, implemente o código abaixo.

```python
# Criando os dados
x = np.random.rand(200)*5
y = x*3 - 1.3 + np.random.randn(200)

# Plotando o gráfico
plt.figure(figsize = (16, 9))
plt.scatter(x, y)
plt.title("Relação entre x e y")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
```

## Exercício 87
#### Objetivo
Criar um gráfico de barra para exibir as vendas de diferentes produtos.

#### Instruções:
- Crie uma lista de produtos (por exemplo, "Produto A", "Produto B", "Produto C").
- Associe um valor de vendas para cada produto.
- Use plt.bar() para criar o gráfico de barras.
- Adicione um título, rótulos para os eixos e personalize as cores das barras.

## Exercício 89
#### Objetivo:
Criar um histograma para visualizar a distribuição de uma variável contínua, como o salário.

#### Instruções:
- Gere uma lista de salários usando uma distribuição aleatória.
- Use plt.hist() para criar o histograma.
- Especifique o número de bins.
- Adicione um título, rótulos para os eixos e personalize a cor do histograma.

```python
np.random.seed(0)
salarios = np.random.exponential(scale = 1300, size = 1000)

plt.figure(figsize = (16, 9))
plt.hist(salarios, bins = 30)
plt.title("Distribuição dos salários")
plt.ylabel("Quantidade de funcionários")
plt.xlabel("Salário")
plt.show()
```

## Exercício 90
#### Objetivo:
Combinar gráficos de linha, barra, dispersão e histograma em um layout de subplot.

#### Instruções:
- Crie uma figura com plt.figure() e especifique um tamanho adequado.
- Use plt.subplot() para criar um layout de 2x2.
- No primeiro subplot, crie um gráfico de linha.
- No segundo subplot, crie um gráfico de barra.
- No terceiro subplot, crie um gráfico de dispersão.
- No quarto subplot, crie um histograma.
- Para cada subplot, adicione um título e rótulos apropriados para os eixos.

```python
plt.figure(figsize = (20, 10))

# Gráfico de linha
plt.subplot(2, 2, 1)
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y, 'g:', label = "Linha")
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title("Seno de X")
plt.legend()


# Gráfico de barras
produtos = ['Produto A', 'Produto B', 'Produto C']
vendas = [10000, 100000, 57800]

plt.subplot(2, 2, 2)
plt.bar(produtos, vendas, fill = True, edgecolor = 'black', label = 'Barras')
plt.title("Vendas para cada produto (em milhões)")
plt.xlabel("Produto")
plt.ylabel("Vendas (em milhões)")
plt.legend()


# Gráfico de dispersão
peso = [67.8, 72.9, 76.4, 89.5, 90.7, 94.8]
altura = [1.59, 1.68, 1.76, 1.69, 1.85, 1.8]

plt.subplot(2, 2, 3)
plt.scatter(peso, altura, color = 'b', marker = '^', label = 'Dispersão')
plt.title("Relação entre peso e altura")
plt.xlabel("Peso")
plt.ylabel("Altura")
plt.legend()

# Histograma
plt.subplot(2, 2, 4)
plt.hist(salarios, bins = 30, label = 'Histograma', color = 'green')
plt.title("Distribuição dos salários")
plt.ylabel("Quantidade de funcionários")
plt.xlabel("Salário")
plt.legend()

plt.tight_layout()
plt.show()
```
